```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
import math
```

```python
%matplotlib inline
```

```python
def chendushow(img):
    img=img.copy()
    if len(img.shape)==2:
        img=cv2.cvtColor(img,cv2.COLOR_GRAY2RGB)
    elif len(img.shape)==3:
        img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    else:
        print(img.shape)
    plt.imshow(img)
```

```python
img_dep=cv2.imread('deep.jpg',0)
img_bgr=cv2.imread('bgr.jpg')
img_bgr_gray=cv2.cvtColor(img_bgr,cv2.COLOR_BGR2GRAY)
```

```python
chendushow(img_dep)
```

```python
print(img_dep.shape,img_dep)
chendushow(img_bgr)
```

```python
chendushow(img_bgr_gray)
```

```python
ret, imag0 = cv2.threshold(img_bgr, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
chendushow(imag0)
print(imag0.shape,imag0)
```

```python
chendushow(imag0)
temp,contours, hierarchy = cv2.findContours(imag0, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

print(len(contours))
n=0

for cnt in contours:
    if len(cnt) > 50:
        try:
            ellipse = cv2.fitEllipse(cnt)
            
        except cv2.error:
            continue
        print(ellipse)    
        S1 = cv2.contourArea(cnt)
        S2 = math.pi * ellipse[1][0] * ellipse[1][1]
        print(S1,S2)
        if (S1 / S2) < 0.4:
            cv2.ellipse(img_bgr, ellipse, (255, 0, 0), 2)
            n += 1
#             print(n,cnt)
chendushow(img_bgr)
```
